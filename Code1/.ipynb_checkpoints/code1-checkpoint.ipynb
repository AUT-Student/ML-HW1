# Polynomial

```python
class Polynomial():
    def __init__(degree):
        self.degree = degree
        self.c = np.zeros(degree+1) # Coefficients of polynomial: c[n]*(x**n) + ... + c[2]*(x**2) + c[1]*(x) + c[0]
    
    def single_apply(self, x):
        sum = 0
        for i, ci in enumerate(self.c):
            sum += ci * (x**i)

        return sum
    
    def apply(self, x):
        return np.array(map(self.single_apply, x))
    
    
    def set_c(self, c):
        self.c = c
    
    def get_c(self, c):
        return c
    
    def plot(self):
        x = np.arange(-1, +1, 0.01)
        y = self.apply(x)

        plt.plot(x, y, color="red")
        plt.scatter(x_train, y_train)
        plt.show()
        
    #     @staticmethod   
#     def single_polynomial(c, x):
#         sum = 0
#         for i, ci in enumerate(c):
#             sum += ci * (x**i)

#         return sum
    
#     @staticmethod
#     def polynomial(c, x):
#         return np.array([single_polynomial(c, xi) for xi in x])
```

# Loss Function

```python
class LossFunction():
    
    @staticmethod
    def mse(y_predict, y_real):
        n = len(y_predict)
        sum = 0
        for i in range(n):
            sum += (y_real[i]-y_predict[i])**2

        return sum/n

    @staticmethod
    def partial_mse(y_predict, y_real, x, partial_degree):
        n = len(x)
        sum = 0
        for i in range(n):
            sum +=  x[i]**partial_degree * (y_real[i] - y_predict[i])
        return -2/n*sum
    
    @staticmethod
    def rmse(y_predict, y_real):
        mse = LossFunction.mse(y_predict, y_real)
        return mse**(0.5)
    
    @staticmethod
    def partial_rmse(y_predict, y_real, x, partial_degree):
        return partial_mse(y_predict, y_real, x, partial_degree) / (2 * LossFunction.rmse(y_predict, y_real))
    
    @staticmethod
    def mae(y_predict, y_real):
        n = len(y_predict)
        sum = 0
        for i in range(n):
            sum += abs(y_real[i]-y_predict[i])
        
        return sum/n
    
    @staticmethod
    def partial_mae(y_predict, y_real, x, partial_degree):
        n = len(x)
        sum = 0
        for i in range(n):
            sum += np.sign(y_real[i]-y_predict[i]) * -1 * (x[i]**partial_degree)

        return sum/n
```

# Gradient Descent

```python
class PolynomialGradientDescent(): 
    def __init__(self, number_epoch, degree, loss_function_name):
        self.learning_rate_initial = 0.5
        self.learning_rate = self.learning_rate_initial
        self.decrease_learning_rate = self.learning_rate_initial / number_epoch
            
        self.train_error = []
        self.test_error = []
        self.learning_rate_value = []
    
        self.number_epoch = number_epoch # Number of required epochs
        
        self.polynomial = Polynomial(degree=degree)
        
        self._set_loss_function(loss_function_name)
            
    def _set_loss_function(self, loss_function_name):
        if loss_function_name == "MSE":
            self.loss_function = LossFuction.mse
            self.partial_loss_function = LossFuction.partial_mse
        elif loss_function_name == "RMSE":
            self.loss_function = LossFuction.rmse
            self.partial_loss_function = LossFuction.partial_rmse
        elif loss_function_name == "MAE":
            self.loss_function = LossFuction.mae
            self.partial_loss_function = LossFuction.partial_mae
        else:
            raise Exception("Unkwon Loss Function Name!")
        
    def run(self):
        for epoch in range(self.number_epoch):
            self._collect_epoch_statistic(epoch)
            
            c_new = []
            y_predict = self.polynomial.apply(x_train)
            
            for i, ci in enumerate(self.polynomial.get_c()):
                ci_new = ci - self.learning_rate * self.partial_loss_function(y_predict, y_train, x_train, i)
                c_new.append(ci_new)

            self.polynomial.set_c(np.array(c_new))
            self.learning_rate -= self.decrease_learning_rate
            
        self._collect_epoch_statistic(epoch)
            
    def _collect_epoch_statistic(self, epoch):
        self.learning_rate_value.append(self.learning_rate)
        
        y_predict = self.polynomial.apply(x_train)
        self.train_error.append(self.loss_function(y_predict, y_train))
        
        y_predict = self.polynomial.apply(x_test)
        self.test_error.append(self.loss_function(y_predict, y_test))
            
    def plot_final_prediction(self):
        self.polynomial.plot()
                
            
    def plot_train_test_error(self):
        x = np.arange(0, self.number_epoch+1)
        
        plt.plot(x, self.train_error, color="green", label="Train Error")
        plt.plot(x, self.test_error, color="red", label="Test Error")
        plt.legend()
        plt.title("Train Error VS Test Error")
        plt.xlabel("Epoch")
        plt.ylabel("Error")
        plt.show()
        
    
    def plot_learning_rate_value(self):
        x = np.arange(0, self.number_epoch+1)
        
        plt.plot(x, self.learning_rate_value, color = "yellow")
        plt.xlabel("Epoch")
        plt.ylabel("Learning Rate")
        plt.title("Learning Rete Changes")
        plt.show()
```

```python
polynomial_gradient_descent = PolynomialGradientDescent(number_epoch = 10000, degree = 100, loss_function_name="MSE")
```

```python
polynomial_gradient_descent.run()
```

```python
polynomial_gradient_descent.plot_final_prediction()
```

```python
polynomial_gradient_descent.plot_train_test_error()
```

# Normal Equation

```python
class PolynomialNormalEquation():
    def __init__(self, degree, lambda_=0):
        self.degree = degree
        self.lambda_ = lambda_
        self.polynomial = Polynomial(degree=degree)
        
    def _calculate_x(self):
        self.x = []
        for xi_train in x_train:
            new_x = []
            for di in range(self.degree+1):
                new_x.append(xi_train**(di))
            self.x.append(new_x)
        
        self.x = np.array(self.x)    
    
    def _normal_equation(self):
        xt = np.transpose(self.x)
        xtx = np.matmul(xt, self.x)
        
        lambda_matrix = np.identity(self.degree + 1) * self.lambda_
        lambda_matrix[0, 0] = 0
        xtx += lambda_matrix
        
        xtxi = np.linalg.inv(xtx)
        xty = np.matmul(xt, y_train)
        c = np.matmul(xtxi, xty)
        self.polynomial.set_c(c)
    
    def plot_final_prediction(self):
        self.polynomial.plot()
        
    def run(self):
        self._calculate_x()
        self._normal_equation()
        
```

```python
polynomial_normal_equation = PolynomialNormalEquation(degree=8, lambda_=1)
```

```python
polynomial_normal_equation.run()
```

```python
polynomial_normal_equation.plot_final_prediction()
```
